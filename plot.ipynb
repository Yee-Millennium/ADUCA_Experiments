```python
import matplotlib.pyplot as plt

def plot_optmeasures(data1, data2):
    iterations_1 = data1['iterations']
    optmeasures_1 = data1['optmeasures']
    iterations_2 = data2['iterations']
    optmeasures_2 = data2['optmeasures']

    plt.figure(figsize=(8, 8))

    plt.plot(iterations_1, optmeasures_1, marker='.', linestyle='-.', color='b', label = 'AdapCODER')
    plt.plot(iterations_2, optmeasures_2, marker='.', linestyle='', color ='r', label = 'CODER_linesearch')

    plt.xlabel('Iterations')
    plt.ylabel('f')

    plt.title('Optmeasures vs Iterations')
    plt.grid(True)
    plt.legend()

    plt.savefig("output/")
    plt.show()
```

```python
import json

# Load the JSON data
# output/a9a-0.0_0.0-AdapCODER-0.0-output-2024-08-12_04-42-11-488756.json (small constants)
# output/a9a-0.0_0.0-AdapCODER-0.0-output-2024-08-12_04-53-24-733206.json (large)
with open('output/a9a-0.0_0.0-AdapCODER-0.0-output-2024-08-12_04-53-24-733206.json') as f1:
    data1 = json.load(f1)
with open('output/a9a-0.0_0.0-CODER_linesearch-0.0-output-2024-08-12_04-45-22-036585.json') as f2:
    data2 = json.load(f2)

plot_optmeasures(data1, data2)
```
