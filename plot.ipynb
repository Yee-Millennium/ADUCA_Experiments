```python
import matplotlib.pyplot as plt
import numpy as np

def plot_optmeasures(data1, data2, data3, optval):
    iterations_1 = np.array(data1['iterations'])
    optmeasures_1 = np.array(data1['optmeasures']) - optval
    iterations_2 = np.array(data2['iterations'])
    optmeasures_2 = np.array(data2['optmeasures']) - optval
    iterations_3 = np.array(data3['iterations'])
    optmeasures_3 = np.array(data3['optmeasures']) - optval

    fig, ax1 = plt.subplots(figsize=(8,8))
    fig.set_dpi(150)
    ax1.set_yscale('log')

    ax1.plot(iterations_1, optmeasures_1, marker='.', linestyle='-.', color='b', label = data1['args']['algo'])
    ax1.plot(iterations_2, optmeasures_2, marker='.', linestyle=':', color ='r', label = data2['args']['algo'])
    ax1.plot(iterations_3, optmeasures_3, marker='.', linestyle='-.', color ='g', label = data3['args']['algo'])
    
    ax1.set_xlabel('Full Gradient Evaluations')
    ax1.set_ylabel(r'$f - f^*$')
    ax1.set_title(f"dataset: {data1['args']['dataset']}")

    fig.savefig(f"output/{data1['args']['dataset']}")

    plt.grid(True)
    plt.legend()
    plt.show()
```
